```python
import sys
import os
project_root = os.path.abspath('..')
sys.path.append(project_root)

import json
from evaluation.baseline_metrics_calc import (
    load_jsonl,
    baseline_coherence_score,
    baseline_relevance_score,
    baseline_redundancy_score, 
    baseline_acceptability_score, 
    baseline_circularity_score
)
import pandas as pd
```

```python
file_path = os.path.join(project_root, "results", "generation", "cot_outputs_ollama_meta_reasoning_conclusion.jsonl")
cot_data = load_jsonl(file_path)

# Evaluate CoT prompting method based on defined metrics
for entry in cot_data:
    exp = entry.get("cot_explanation", "")
    print("Question:", entry["question"])
    print("Coherence:", baseline_coherence_score(exp))
    print("Relevance:", baseline_relevance_score(exp))
    print("Redundancy:", baseline_redundancy_score(exp))
    print()
```

```python
file_path = os.path.join(project_root, "results", "generation", "argllm_outputs_ollama.jsonl")
argllm_data = load_jsonl(file_path)

# Evaluate argLLM method based on defined metrics
results = []
for entry in argllm_data:
    question = entry.get("question", "")
    base_bag = entry["base"]["bag"]
    estimated_bag = entry["estimated"]["bag"]

    results.append({
        "question": question,
        "base_acceptability": baseline_acceptability_score(base_bag),
        "base_circularity": baseline_circularity_score(base_bag),
        "estimated_acceptability": baseline_acceptability_score(estimated_bag),
        "estimated_circularity": baseline_circularity_score(estimated_bag),
    })

# Display in DataFrame
df = pd.DataFrame(results)
print(df)   
```
